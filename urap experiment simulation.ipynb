```python
#Import the required modules
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tkinter import *
import time
```

```python
#Experimnent 1 simulation
#Initialize the tkinter window and canvas, and the response array
master = Tk()
master.geometry('2560x1600')
master.config(bg='black')
can = Canvas(master, width=2560, height=1600, bg='black')
can.pack()
response_dict = {}
num_sim = 0
num_trial = 0
w = master.winfo_screenwidth()
h = master.winfo_screenheight()
xsize = 100
ysize = 250
x0 = (w/2)-(xsize/2)
x1 = (w/2)+(xsize/2)
y0 = (h/2)-(ysize/2)

#Define the required functions
def rect():
    global lot_type
    lot_type = np.random.choice(['risk', 'ambiguous'])
    amt = np.random.choice([50, 95, 180, 340, 650], 2)
    win_amt, lose_amt = amt[0], amt[1]
    if lot_type == 'risk':
        prob = np.random.choice([0.13, 0.25, 0.38])
        y1 = y0+(ysize*prob)
        can.create_rectangle(x0, y0, x1, y1, fill="red")
        can.create_rectangle(x0, y1, x1, y1+(250*(1-prob)), fill="blue")
        can.create_text(x0+50, y0-15, text=win_amt, fill="white")
        can.create_text(x0+50, y0+ysize+15, text=lose_amt, fill="white")
    elif lot_type == 'ambiguous':
        occluder = np.random.choice([0.25, 0.5, 0.75])
        y1 = y0+ysize/2
        can.create_rectangle(x0, y0, x1, y1, fill="red")
        can.create_rectangle(x0, y1, x1, y0+ysize, fill="blue")
        can.create_rectangle(x0-10, y0+(ysize/2*occluder), x1+10, y1+(ysize/2*occluder), fill="grey")
        can.create_text(x0+50, y0-15, text=win_amt, fill="white")
        can.create_text(x0+50, y0+ysize+15, text=lose_amt, fill="white")

def circle(col):
    can.create_oval((w/2)-5, (h/2)-5, (w/2)+5, (h/2)+5, fill=col)

def response(button):
    response_dict[lot_type+' '+str(num_trial)] = button

def button():
    global ref_btn, lot_btn
    ref_btn = Button(master, bg='yellow', text = 'Reference Lottery', bd=5, command=response('reference'))
    ref_btn.place(x=x0, y=y0-25)
    lot_btn = Button(master, bg='white', text = 'Task Lottery', bd=5, command=response('trial'))
    lot_btn.place(x=x0, y=y0+25)

def delete(par):
    can.delete(par)

def set_programState(new_state, trial=num_trial, sim=num_sim):
    global programState
    programState = new_state
    if new_state == 1:
        sim += 1
        trial = 0
    else:
        trial += 1
    if sim < 2:
        run_simulation()
    else:
        can.create_text(w/2, h/2, text='End of Experiment 1!', fill="white")

def run_simulation():
    if programState == 1:
        can.after(0, delete, 'all')
        can.after(0, rect)
        can.after(3000, set_programState, 2)
    elif programState == 2:
        can.after(0, delete, 'all')
        can.after(0, circle, 'white')
        can.after(5000, set_programState, 3)
    elif programState == 3:
        can.after(0, delete, 'all')
        can.after(0, circle, 'green')
        can.after(200, set_programState, 4)
    elif programState == 4:
        can.after(0, delete, 'all')
        can.after(0, button)
        can.after(1000, set_programState, 5)
    elif programState == 5:
        ref_btn.destroy()
        lot_btn.destroy()
        can.after(0, circle, 'white')
        can.after(4000, set_programState, 1)
        
#Run the simulation
set_programState(1, num_trial, num_sim)
master.mainloop()
```
