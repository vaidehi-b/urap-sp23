```python
#Import the required modules
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tkinter import *
import time
```

```python
#Experimnent 1 simulation
#Initialize the tkinter window and canvas, and the response array
master = Tk()
master.geometry('400x600')
master.config(bg='black')
can = Canvas(master, width=400, height=600, bg='black')
can.pack()
response_array = []
count = 0

#Define the required functions
def rect():
    lot_type = np.random.choice(['risk', 'ambiguous'])
    amt = np.random.choice([50, 95, 180, 340, 650], 2)
    win_amt, lose_amt = amt[0], amt[1]
    if lot_type == 'risk':
        prob = np.random.choice([0.13, 0.25, 0.38])
        y1 = 100+(240*prob)
        can.create_rectangle(100, 100, 200, y1, fill="red")
        can.create_rectangle(100, y1, 200, y1+(240*(1-prob)), fill="blue")
        can.create_text(150, 90, text=win_amt, fill="white")
        can.create_text(150, 330, text=lose_amt, fill="white")
    elif lot_type == 'ambiguous':
        occluder = np.random.choice([0.25, 0.5, 0.75])
        can.create_rectangle(100, 100, 200, 220, fill="red")
        can.create_rectangle(100, 200, 200, 320, fill="blue")
        can.create_rectangle(90, 100+(240*occluder/2), 210, 320-(240*occluder/2), fill="grey")
        can.create_text(150, 90, text=win_amt, fill="white")
        can.create_text(150, 330, text=lose_amt, fill="white")        

def circle(col):
    if col == 'w':
        can.create_oval(200, 200, 210, 210, fill='white')
    elif col == 'g':
        can.create_oval(200, 200, 210, 210, fill='green')

def response(button_id):
    if button_id == 1:
        response_array.append('ref')
    elif button_id == 2:
        response_array.append('lot')

def button():
    global ref_btn, lot_btn
    ref_btn = Button(master, bg='yellow', text = 'Reference Lottery', bd=5, command=response(1))
    ref_btn.place(x=100, y=200)
    lot_btn = Button(master, bg='white', text = 'Task Lottery', bd=5, command=response(2))
    lot_btn.place(x=100, y=250)

def delete(par):
    can.delete(par)

def set_programState(new_state, count=count):
    global programState
    programState = new_state
    if new_state == 1:
        count += 1
    if count < 11:
        run_simulation()

def run_simulation():
    if programState == 1:
        can.after(0, delete, 'all')
        can.after(0, rect)
        can.after(3000, set_programState, 2)
    elif programState == 2:
        can.after(0, delete, 'all')
        can.after(0, circle, 'w')
        can.after(5000, set_programState, 3)
    elif programState == 3:
        can.after(0, delete, 'all')
        can.after(0, circle, 'g')
        can.after(200, set_programState, 4)
    elif programState == 4:
        can.after(0, delete, 'all')
        can.after(0, button)
        can.after(1000, set_programState, 5)
    elif programState == 5:
        ref_btn.destroy()
        lot_btn.destroy()
        can.after(0, circle, 'w')
        can.after(4000, set_programState, 1)
        
#Run the simulation
set_programState(1)
master.mainloop()
```
